1.- Programa en python que genere un número entre el 1 y 10 (no visible)  preguntarle al usuario que número cree que generó la computadora, el usuario tendrá 3 oportunidades de adivinar.

Decir si adivino o si falló en sus 3 intentos.



NOTA: 100% VALIDADO (usar función para validar números)



NOTA2: El usuario podrá jugar cuantas veces lo desee, al final del juego desplegar cantidad de ganados y perdidos



```python
#Felix Peraza Luis Zahir 369453
#12/04/24
#Programa que genere un numero no visible entre el 1 y el 10, y que el usuario trate de adivinar que numero es
#adivina num

import random

def vali_num(num):
  try:
    num = int(num)
    if 1 <= num <= 10:
      return True
    else:
      return False
  except ValueError:
    return False

def game():
  intentos = 3
  num_generado = random.randint(1,10)

  print("Adivina el numero que ha generado la computadora entre 1 y 10")

  while intentos > 0:
    guess = input("Adivina que numero sera el generado: ")
    if vali_num(guess):
      guess = int(guess)
      if guess == num_generado:
        print("Haz adivinado +1 punto")
        return True
      else:
        intentos -= 1
        print(f"Incorrecto. Te quedan {intentos} intentos")
    else:
      print("ERROR: Ingrese un numero entero valido entre 1 y 10")
  print(f"El numero era {num_generado}, Ya se te acabaron las vidas")
  return False

win = 0
lose = 0

while True:
  if game():
    win += 1
  else:
    lose += 1
  new_game = input("Quieres seguir jugando? (s/n)")
  if new_game.lower() != "s":
    break

print(f"Total de juegos ganados: {win}")
print(f"Total de jugos perdidos: {lose}")
```

```python
import random

def generar_lista_numeros():
    numeros = random.sample(range(1, 11), 10)
    return numeros

def jugar_busca_numero():
    lista_numeros = generar_lista_numeros()
    intentos_maximos = 3
    numero_a_adivinar = random.choice(lista_numeros)

    print("Bienvenido al juego Busca Número!")
    print("Se ha generado una lista de 10 números del 1 al 10.")
    print("Tienes que adivinar en qué posición se encuentra el número.")

    for intento in range(intentos_maximos):
        print(f"\nIntento {intento + 1}/{intentos_maximos}:")
        indice = int(input("Ingresa el índice donde crees que está el número: "))

        if indice < 0 or indice >= len(lista_numeros):
            print("Índice fuera de rango. Inténtalo de nuevo.")
            continue

        if lista_numeros[indice] == numero_a_adivinar:
            print("¡GANASTE! El número estaba en la posición correcta.")
            return
        else:
            print("¡Incorrecto!")
            if intento < intentos_maximos - 1:
                print("Tienes un nuevo intento.")

    print(f"\nLo siento, has perdido. El número era {numero_a_adivinar}.")

while True:
    jugar_busca_numero()
    continuar = input("¿Quieres jugar de nuevo? (s/n): ")
    if continuar.lower() != 's':
        break
```
