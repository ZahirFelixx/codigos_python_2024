```python
def menu():
    if opc == 1:
        numeros()

    elif opc == 2:
        impares()

    elif opc == 3:
        rango()

    elif opc == 5:
        promedio()

    else:
        print("El valor ingresado no existe, intente de nuevo...")

def numeros():

    suma = 0
    total = 0
    sumanumeros = []

    while True:
        num = int(input("Ingrese un numero o ingrese 0 para terminar: "))
        if num == 0:
            break

        sumanumeros.append(num)
        suma += num
        total += 1

        media = suma / total
        maxi = max(sumanumeros)
        mini = min(sumanumeros)

        print(f"La suma total de los numeros es {suma}")
        print(f"La media de los numeros es {media}")
        print(f"El valor maximo es {maxi}")
        print(f"El valor minimo es {mini}")
    menu()

def impares():
    import random
    impar = 0
    for i in range(25):
        x = random.randint(10,60)
        if x % 2 != 0:
            print(x)

            impar += 1

            if impar == 15:
                break

    else:
        print("Los 25 numeros se han completado")
        menu()


print("Menu multivariable, ingrese la opcion que desee")


try:
    opc = int(input("1-.numeros 2-. impares 3-. rango 4-. triangulo 5-. promedio"))
    menu()
except ValueError:
    print("Error: Solo se permiten numeros ")

def promedio():
    try:
        calf1 = float(input("Ingrese la calificacion del primer parcial"))
        calf2 = float(input("Ingrese la calificacion del segundo parcial"))
        calf3 = float(input("Ingrese la calificacion del tercer parcial"))
        calf4 = float(input("Ingrese la calificacion del cuarto parcial"))

        prom = 0
        prom = (calf1 + calf2 + calf3 + calf4) / 4

        if prom > 60:
            print("Felicidades, estas aprobado ")
            print("Te esperamos en el siguiente semestre")

        elif prom < 60:
            print("¿Cuantas veces a reprobado?")
            repro = int(input("Ingrese 1, 2, o 3"))
            if repro == 1:
                print("Tienes 2 opprtunidades de repetir")
            elif repro == 2:
                print("Esta es tu ultima oportunidad de repetir")
            elif repro == 3:
                print("Haz reprobado por 3 ocaciones, seras dado de baja academica")

        menu()

    except ValueError:
        print("Error: Los valores ingresados deben ser números enteros.")
    except:
        print("Ocurrió un error inesperado.")

def rango():

    numeros = []

    while True:
        inicio = int(input("Ingrese el inicio del rango: "))
        fin = int(input("Ingrese el final del rango: "))

        for num in range(inicio, fin + 1):
            numeros.append(num)

        cant_numeros = len(numeros)
        suma = sum(numeros)
        promedio = suma / cant_numeros

        print(f"Se generaron {cant_numeros} numeros.")
        print(f"La suma de los numeros es: {suma}")
        print(f"El promedio de los numeros es: {promedio}")
menu()

```
